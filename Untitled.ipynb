```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
p = np.arange(0,1,0.01)
```

```python
plt.plot(p,-1 * np.log(p))
plt.show()
```

```python
plt.plot(p,-1 * (np.log(1- p)))
plt.show()
```

```python
plt.plot(p,-1 * np.log(p))
plt.plot(p,-1 * (np.log(1- p)), 'r')
plt.show()
```
